### Visually check the colors of the clothes

```python
for target in color['major_color'].unique():
    print('Major color:', target)
    
    df = color[color['major_color']==target]
    position = [0, df.shape[0]//3, df.shape[0]*2//3, df.shape[0]-1]

    fig, axes = plt.subplots(1,4, figsize=(20,16))
    for x in range(4):
        path = 'data/In-shop/' + df.iloc[position[x],0]
        img = cv2.imread(path)
        axes[x].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

### Cropping images

```python
color.set_index('image_name', drop=True, inplace=True)
```

```python
color
```

```python
crop_df = pd.DataFrame(columns=['image_name', 'image_color'])
crop_df
```

```python
for x in range(bbox.shape[0]):
    path = 'data/In-shop/' + bbox.iloc[x]['image_name']
    img = cv2.imread(path)
    img_color = color.loc[bbox.iloc[x]['image_name'], 'labelled_color']
        
    if (bbox.iloc[x,6]-bbox.iloc[x,4] >=50) & (bbox.iloc[x,5]-bbox.iloc[x,3] >=50):
        crop = img[bbox.iloc[x,4]:bbox.iloc[x,6], bbox.iloc[x,3]:bbox.iloc[x,5]].copy()
        cv2.imwrite(f'data/In-shop/cropped images/img_{x}.jpg', crop)
        crop_df = crop_df.append({'image_name':f'img_{x}.jpg', 'image_color':img_color}, ignore_index=True)
```

```python
crop_df.to_csv('cropped_image.csv')
```
